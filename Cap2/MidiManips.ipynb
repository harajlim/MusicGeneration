```python
test=converter.parse('Classical/beethoven_opus10_1.mid')
```

```python

parts = test.flat.notes
```

```python
len(parts)
```

```python
notes=[]
durs=[]
offsets=[]
for notehere in parts:

    if isinstance(notehere,note.Note):
        notes.append(str(notehere.pitch))
        durs.append(notehere.duration.quarterLength)
        offsets.append(notehere.offset)
    elif isinstance(notehere,chord.Chord):
        notes.append('.'.join(str(n) for n in notehere.normalOrder))
        durs.append(notehere.duration.quarterLength)
        offsets.append(notehere.offset)
        
```

```python
for i,dur in enumerate(durs):
    print('start at {}'.format(offsets[i]))
    print('last for :{}'.format(dur))
    
```
